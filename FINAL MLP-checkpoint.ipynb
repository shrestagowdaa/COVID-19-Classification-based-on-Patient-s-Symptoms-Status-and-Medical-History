```python
df = df.drop_duplicates()
df
```

```python
# Resetting index
df.reset_index(drop=True, inplace=True)
df=df.dropna()
```

```python
# Chosing target feature
target = "CLASIFFICATION_FINAL"

# Splitting data
X = df.drop(target, axis=1)
y = df[target]

# Printing X
X
```

```python
y
```

```python
X["DATE_DIED"].value_counts()
```

```python
# Calculating percent of missing data
X["DATE_DIED"].value_counts()[0] / X["DATE_DIED"].value_counts().sum()
```

```python
X.drop("DATE_DIED", axis=1, inplace=True)

# Printing data
X.head()
```

```python
y.value_counts()
```

```python
# Transforming multiclass CLASIFFICATION_FINAL feature into binary feature
y = y.apply(lambda x: 1 if x <= 3 else 0)

# Printing target feature value counts
y.value_counts()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Calcualting % of classes in target features
percantages = np.round(y.value_counts() / X.shape[0], 2)

# Barplot
sns.barplot(x=y, y=percantages);
```

```python
from imblearn.under_sampling import RandomUnderSampler

# RandomUnderSampler
X, y = RandomUnderSampler().fit_resample(X, y)

# Resetting index
df.reset_index(drop=True, inplace=True)

# Barplot
sns.countplot(x=y);
```

```python
# Getting all discrete type features
categorical_features = X.select_dtypes(include = "int64").columns.drop("AGE")

# Printing all discrete features
categorical_features
```

```python
from sklearn.preprocessing import LabelEncoder

# Trasforming features using LabelEncoder
X[categorical_features] = df[categorical_features].apply(LabelEncoder().fit_transform)

# Prting new features
X
```

```python
from sklearn.preprocessing import RobustScaler

# Scaling feature
X["AGE"] = RobustScaler().fit_transform(X[["AGE"]])

# Printing new feature
X.AGE
```

```python
X.hist(figsize=(20,20));
```

```python
# Creating copy of X data
full_data = X.copy()

# Adding target column to copy of X
full_data["target"] = y

# Computing the correlation matrix
corr_matrix = np.round(np.corrcoef(full_data, rowvar=False), 2)

# Plotting the correlation matrix
plt.figure(figsize=(14,11))
sns.heatmap(corr_matrix, cmap='viridis', annot=True,
            xticklabels=full_data.columns, yticklabels=full_data.columns
           )
plt.title("Correlation Matrix");
```

```python
from sklearn.decomposition import PCA


# Transforming data into 2dim using PCA
pca = PCA(2)
pca_data = pca.fit_transform(X)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

# Evaluating SVM model using Bootstraping method
scores = cross_val_score(SVC(), X, y)

# Calculating mean of scores
scores.mean()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Evaluating GradientBoostingClassifier model using Bootstraping method
scores = cross_val_score(GradientBoostingClassifier(), X, y)

# Calculating mean of scores
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
# Evaluating the model using cross-validation
scores = cross_val_score(RandomForestClassifier(), X, y)
scores.mean()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
scores = cross_val_score(LogisticRegression(), X, y)
scores.mean()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Create instances of the models
lr_model = LogisticRegression()

# Perform cross-validation and obtain predicted labels
lr_pred = cross_val_predict(lr_model, X, y)

# Calculate confusion matrix
cm = confusion_matrix(y, lr_pred)
print("Confusion Matrix:")
print(cm)

# Calculate accuracy
accuracy = accuracy_score(y, lr_pred)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(y, lr_pred)
print("Precision:", precision)

# Calculate recall
recall = recall_score(y, lr_pred)
print("Recall:", recall)

# Calculate F1 score
f1 = f1_score(y, lr_pred)
print("F1 Score:", f1)
```
